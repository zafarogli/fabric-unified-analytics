```python
df = spark.read.table("silver_taxi_trips")
```

```python
from pyspark.sql.functions import count, sum, avg

taxi_daily = (
    df.groupBy("trip_date")
      .agg(
          count("*").alias("trip_count"),
          sum("fare_amount").alias("total_fare_usd"),
          avg("fare_amount").alias("avg_fare_usd"),
          avg("trip_distance").alias("avg_trip_distance")
      )
)
```

```python
taxi_daily.show()
```

```python
(
    taxi_daily.write
    .format("delta")
    .mode("overwrite")
    .saveAsTable("gold_taxi_daily")
)
```
