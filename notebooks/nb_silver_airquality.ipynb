```python
from pyspark.sql.functions import col, to_date, lit

bronze_air = spark.read.table("bronze_openaq")  # agar table nomi boshqacha bo'lsa, shuni o'zgartirasan

silver_air = (
    bronze_air
    .withColumn("air_date", to_date(col("period.datetimeFrom.local")))
    .select(
        col("air_date"),
        col("location_id").cast("long").alias("location_id"),
        col("sensor_id").cast("long").alias("sensor_id"),
        col("latitude").cast("double").alias("latitude"),
        col("longitude").cast("double").alias("longitude"),
        col("location_name").alias("location_name"),
        col("summary.avg").cast("double").alias("pm25_avg"),
        col("summary.min").cast("double").alias("pm25_min"),
        col("summary.max").cast("double").alias("pm25_max"),
        col("coverage.percentCoverage").cast("double").alias("coverage_pct")
    )
    .filter(col("air_date").isNotNull())
)
```

```python
(
    silver_air.write
    .format("delta")
    .mode("overwrite")
    .saveAsTable("silver_airquality_daily")
)
```
