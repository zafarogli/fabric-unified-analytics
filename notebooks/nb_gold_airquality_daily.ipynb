```python
from pyspark.sql.functions import col, avg

air = spark.read.table("silver_airquality_daily")

gold_airquality_daily = (
    air.groupBy("air_date")
       .agg(
           avg(col("pm25_avg")).alias("pm25_avg_nyc"),
           avg(col("pm25_min")).alias("pm25_min_nyc"),
           avg(col("pm25_max")).alias("pm25_max_nyc"),
           avg(col("coverage_pct")).alias("avg_coverage_pct")
       )
       .filter(col("air_date").isNotNull())
)
```

```python

gold_airquality_daily.show(10)
```

```python
(
    gold_airquality_daily.write
    .format("delta")
    .mode("overwrite")
    .saveAsTable("gold_airquality_daily")
)
```
