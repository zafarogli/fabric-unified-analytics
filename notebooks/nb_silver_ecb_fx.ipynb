```python
from pyspark.sql.functions import col, to_date

bronze_fx = spark.read.table("bronze_ecb_fx")

silver_fx = (
    bronze_fx
    .select(
        to_date(col("TIME_PERIOD")).alias("fx_date"),
        col("OBS_VALUE").cast("double").alias("usd_eur_rate")
    )
    .filter(col("fx_date").isNotNull())
)

silver_fx.printSchema()
silver_fx.show(5)
```

```python
(
    silver_fx.write
    .format("delta")
    .mode("overwrite")
    .saveAsTable("silver_fx_daily")
)
```
