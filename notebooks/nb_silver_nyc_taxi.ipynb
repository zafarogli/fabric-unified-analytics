```python
from pyspark.sql.functions import col, to_date

df = spark.read.table("bronze_taxi_trips")

silver_df = (
    df
    .withColumn("pickup_datetime", col("tpep_pickup_datetime").cast("timestamp"))
    .withColumn("dropoff_datetime", col("tpep_dropoff_datetime").cast("timestamp"))
    .withColumn("trip_date", to_date(col("tpep_pickup_datetime")))
    .drop("tpep_pickup_datetime", "tpep_dropoff_datetime")
)

silver_df = silver_df.filter(
    (col("trip_date") >= "2024-01-01") & (col("trip_date") <= "2024-02-01") &
    (col("trip_distance") > 0) & (col("fare_amount") > 0)
)

(
    silver_df.write
    .format("delta")
    .mode("overwrite")
    .saveAsTable("silver_taxi_trips")
)
```
