```python
from pyspark.sql.functions import col

taxi_daily = spark.read.table("lh_unified_analytics.dbo.gold_taxi_daily")
air_daily = spark.read.table("lh_unified_analytics.dbo.gold_airquality_daily")
```

```python
gold_mobility_air = (
    taxi_daily.alias("td")
    .join(
        air_daily.alias("ad"),
        col("td.trip_date") == col("ad.air_date"), "inner"
    )
    .select(
        col("td.trip_date").alias("date"),
        col("td.trip_count"),
        col("td.total_fare_usd"),
        col("td.avg_fare_usd"),
        col("td.avg_trip_distance"),
        col("ad.pm25_avg_nyc"),
        col("ad.pm25_min_nyc"),
        col("ad.pm25_max_nyc"),
        col("ad.avg_coverage_pct")
    )
)
```

```python
gold_mobility_air.show(25)
```

```python
(
    gold_mobility_air.write
    .format("delta")
    .mode("overwrite")
    .saveAsTable("gold_mobility_air_daily")
)
```
