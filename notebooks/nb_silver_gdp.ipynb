```python
from pyspark.sql.functions import col

bronze_gdp = spark.read.table("bronze_worldbank_gdp")

silver_gdp = (
    bronze_gdp
    .select(
        col("countryiso3code").alias("country_code"),
        col("date").cast("int").alias("gdp_year"),
        col("value").cast("double").alias("gdp_usd")
    )
    .filter(col("gdp_usd").isNotNull())
)
```

```python
(
    silver_gdp.write
    .format("delta")
    .mode("overwrite")
    .saveAsTable("silver_gdp_yearly")
)
```
